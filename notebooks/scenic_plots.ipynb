```python
import os
from pathlib import Path
import numpy as np, pandas as pd, scanpy as sc, loompy as lp, anndata as ad
```

```python
# set working directory
wdir = '.'
os.chdir( wdir )

PATH_TO_BIN_MTX = Path("../data/scenic_protocol/files/bin_mtx.csv")
HPCS_META = Path("../data/processed/hpcs_lps_metadata.csv")
```

```python
bin_mtx = pd.read_csv(PATH_TO_BIN_MTX, index_col=0)
metadata = pd.read_csv(HPCS_META, index_col=0)
```

```python
import seaborn as sns
```

```python
bw_palette = sns.xkcd_palette(["white","black"])
```

Get regulon order.

```python
g_hpcs = sns.clustermap(
    bin_mtx,
    cmap=bw_palette
)
# g_hpcs.ax_heatmap.set_xticklabels([])
# g_hpcs.ax_heatmap.set_xticks([])
# g_hpcs.ax_heatmap.set_xlabel('Cells')
# g_hpcs.ax_heatmap.set_ylabel('Regulons')
# g_hpcs.ax_row_dendrogram.set_visible(False)
# g_hpcs.ax_col_dendrogram.set_visible(False)
# g_hpcs.cax.set_visible(False)
```

Somatotropes.

```python
g_som_health = sns.clustermap(
    bin_mtx[metadata[(metadata['cell_type_brief']=="Som") & (metadata['state']=="Healthy")].index].iloc[g_hpcs.dendrogram_row.reordered_ind,],
    cmap=bw_palette,
    row_cluster=False,
    figsize=(9.21,10)
)
g_som_health.ax_heatmap.set_xticklabels([])
g_som_health.ax_heatmap.set_xticks([])
g_som_health.ax_heatmap.set_yticklabels([])
g_som_health.ax_heatmap.set_yticks([])
g_som_health.ax_row_dendrogram.set_visible(False)
g_som_health.ax_col_dendrogram.set_visible(False)
g_som_health.cax.set_visible(False)
g_som_health.fig.savefig(Path('../figures/Fig1/F1i/som_healthy.tiff'), format='tiff', figsize=(9,10))
```

```python
g_som_infl = sns.clustermap(
    bin_mtx[metadata[(metadata['cell_type_brief']=="Som") & (metadata['state']=="Inflammation")].index].iloc[g_hpcs.dendrogram_row.reordered_ind,],
    cmap=bw_palette,
    row_cluster=False,
    figsize=(14.12,10)
)
g_som_infl.ax_heatmap.set_xticklabels([])
g_som_infl.ax_heatmap.set_xticks([])
g_som_infl.ax_heatmap.set_yticklabels([])
g_som_infl.ax_heatmap.set_yticks([])
g_som_infl.ax_row_dendrogram.set_visible(False)
g_som_infl.ax_col_dendrogram.set_visible(False)
g_som_infl.cax.set_visible(False)
g_som_infl.fig.savefig(Path('../figures/Fig1/F1i/som_inflammation.tiff'), format='tiff')
```
