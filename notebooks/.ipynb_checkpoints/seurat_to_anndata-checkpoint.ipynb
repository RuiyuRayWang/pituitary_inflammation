```python
import os
import anndata as ad, scanpy as sc, numpy as np, pandas as pd
from scipy import io
from scipy.sparse import coo_matrix, csr_matrix
```

```python
# load sparse matrix:
X = io.mmread("../data/scvelo/counts.mtx")
```

```python
# create anndata object
adata = ad.AnnData(
    X = X.transpose().tocsr()
)
```

```python
# load cell metadata:
cell_meta = pd.read_csv("../data/scvelo/metadata.csv")
```

```python
# load gene names:
with open("../data/scvelo/gene_names.csv", 'r') as f:
    gene_names = f.read().splitlines()
```

```python
# set anndata observations and index obs by barcodes, var by gene names
adata.obs = cell_meta
adata.obs.index = adata.obs['cell_id']
adata.var.index = gene_names
```

```python
# load dimensional reduction
pca = pd.read_csv('../data/scvelo/pca.csv')
pca.index = adata.obs.index
```

```python
# set pca and umap
adata.obsm['X_pca'] = pca.to_numpy()
adata.obsm['X_umap'] = np.vstack((adata.obs['UMAP_1'].to_numpy(), adata.obs['UMAP_2'].to_numpy())).T
```

```python
# plot a UMAP colored by sampleID to test:
sc.pl.umap(adata, color=['cell_type_brief'], frameon=False, save=True)
```

```python
# save dataset as anndata format
adata.write('../data/scvelo/cells.h5ad')

## BUG: h5py compatibility issue. see https://github.com/theislab/scanpy/issues/1275
```
